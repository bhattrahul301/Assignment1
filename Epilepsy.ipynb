```python
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import confusion_matrix,classification_report
```

```python
X_data =  rdata.loc[:, rdata.columns != 'status']
Y_data = rdata['status']
```

```python
ss = StandardScaler()
X_data = ss.fit_transform(X_data)
```

```python
X_data[0]
```

```python
pca = PCA(n_components=0.95)
pca.fit(X_data)
X_data = pca.transform(X_data)
X_data
```

```python
sns.scatterplot(x=X_data[:,0],y = X_data[:,1], hue=rdata['status'])
```

```python
sns.scatterplot(x=X_data[:,1],y = X_data[:,2], hue=rdata['status'])
```

```python
sns.scatterplot(x=X_data[:,0],y = X_data[:,3], hue=rdata['status'])
```

```python
sns.scatterplot(x=X_data[:,0],y = X_data[:,6], hue=rdata['status'])
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_data,Y_data,random_state=32,test_size=0.25)
```

```python
Y_train.value_counts(),Y_test.value_counts()
```

```python
lr = LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
lr.score(X_test,Y_test)
```

```python
svc = SVC()
svc.fit(X_train,Y_train)
svc.score(X_test,Y_test)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,Y_train)
dt.score(X_test,Y_test)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
rf.score(X_test,Y_test)
```

```python
scores = []
for i in range(1,11):
    knnc = KNeighborsClassifier(n_neighbors=i)
    knnc.fit(X_train,Y_train)
    scores.append(knnc.score(X_test,Y_test))
```

```python
sns.lineplot(x=range(1,11),y = scores)
```

```python
knnc = KNeighborsClassifier(n_neighbors=5)
knnc.fit(X_train,Y_train)
knnc.score(X_test,Y_test)
```

```python
gnb = GaussianNB()
gnb.fit(X_train,Y_train)
gnb.score(X_test,Y_test)
```

```python
pred = knnc.predict(X_test)
confusion_matrix(Y_test,pred)
```
